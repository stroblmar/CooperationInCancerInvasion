# Simulation Analysis of the Cooperative Invasion Model in *Strobl et al (2019)*

This notebook contains the sample simulations presented in our paper, and illustrates how to use the code

```python
# First load the required libraries
import numpy as np
import pandas as pd
import scipy.integrate
import math
import sys
import os
import explicitModelSolver as mySolver # Import solver
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import time
import pickle

# Format plot
sns.set(style="white",
        font_scale=1.5,
        rc={'figure.figsize':(12,6), 
            'lines.linewidth': 3})
```

```python
# Parameterise solver
paramsList=[1.5, .5, .5, 1., 4e-5, 70., 12.5, 10., 0.] # cS, cMA, cAM, rT, DT, rL, dS, dM, eps
domainSizeList = [0, 1, 50.]
dt = 5e-3
dx = 5e-3
method = 'BDF' 
stabiliseSpatDiscB = True
rtol = 1e-10
atol = 1e-10
```

**Note:** As mentioned in Section 4. of the paper we use a numerical stabilisation scheme in order to tackle numerical oscilliations during the solution process. For details of this scheme consult the code in 'modelSolver.py'. In addition, during the development of the code we experimented with a relaxation of the (1-M) term in the flux expressions of the tumour equations (the 'eps' variable in the 'paramsList' array). For all analysis presented in the paper, this term was removed by setting 'eps=0' (as defined above).

### Code for generating publication figures

```python
def GeneratePublicationFigure(model, timesToPlotList, saveFigB=False, figsize=None, 
                              lineWidth=3, ylimVec=[0, 1.1], outName=None):
    tIdxList = [np.argwhere(np.abs(model.loggedTimeVec - tVal) < (model.dx / 2))[0][0] for tVal in timesToPlotList]
    nPlots = len(tIdxList)
    fig, axList = plt.subplots(3,1,sharex=True,figsize=figsize)
    for pltIdx in range(nPlots):
        tIdx = tIdxList[pltIdx]
        currAx = axList[pltIdx]
        currAx.fill_between(model.xVec, np.zeros_like(model.M[:, tIdx]), model.M[:, tIdx],
                         color='#B9B9BA', lw=lineWidth, alpha=0.6, hatch='..',
                         label="ECM")
        currAx.fill_between(model.xVec, np.zeros_like(model.S[:, tIdx]), model.S[:, tIdx],
                         color='#8079FE', lw=lineWidth, alpha=0.6, hatch='//',
                         label="Stroma")
        currAx.fill_between(model.xVec, np.zeros_like(model.A[:, tIdx]), model.A[:, tIdx],
                         color='#B676AA', lw=lineWidth, alpha=0.5, hatch='**',
                         label="Acid")
        currAx.fill_between(model.xVec, np.zeros_like(model.TA[:, tIdx]), model.TA[:, tIdx],
                         color='#FD676B', lw=lineWidth, alpha=0.6, hatch='xx',
                         label=r"$T_A$")
        currAx.fill_between(model.xVec, np.zeros_like(model.TM[:, tIdx]), model.TM[:, tIdx],
                         color='#DDDA6D', lw=lineWidth, alpha=0.6, hatch='++',
                         label=r"$T_M$")
#         if pltIdx==2: currAx.set_xlabel(r"$x$")
#         currAx.set_ylabel("Non-Dimensionalised \n Variables")
        currAx.set_title(r"$t$ = %1.2f" % model.loggedTimeVec[tIdx])
        currAx.set_ylim(ylimVec)
    plt.tight_layout()
    # Save the figure if requested
    if saveFigB:
        if outName == None:
            outName = "t_%1.2f_dt_%1.3f_dx_%1.3f.pdf" % (model.timesToPlotList[tIdx], model.dt, model.dx)
        currAx.savefig(outName, orientation='portrait', format='pdf')
        currAx.close()
```

## 1. Introduction to the Solver Class

I've written my solver as a class, so to have all the parameters and options nicely packaged together. To use it we first create an instance.

```python
m = mySolver.fullModelSolver(paramsList=paramsList, domainSizeList=domainSizeList,
                             dx=dx, dt=dt, stabiliseSpatDiscB=stabiliseSpatDiscB)
```

We have the following options when instantiating it:

Next, you initialise the system. You do so as follows:

```python
m.Set_ICParameters(stromaParamList=[0.2,0.1],tumourAParamList=[0.1,0.1],
               tumourMParamList=[0.1,0.1],matrixParamList=[0.2,0.1],
               acidParamList=[0.1,0.1])
m.ApplyInitialConditions()
```

You can plot the current state using the classes Plot() method.

```python
m.Plot([0])
```

Finally, to run it, we employ the Run() method.

```python
m.Run(method=method, rtol=rtol, atol=atol, logInterval=1.)
```

And we can plot the results as follows:

```python
m.Plot([0,m.tEnd/2,m.tEnd],figsize=(8.27, 11.69),ylimVec=[-0.5,1.3])
```

### Compute edge position over time

```python
# Compute the position of TA over time
edgePosList_TA = []
for tIdx,t in enumerate(m_MonoTA.loggedTimeVec):
    edgePosList_TA.append(m_MonoTA.xVec[np.argmax(np.abs(np.gradient(m_MonoTA.TA[:,tIdx])))])

# Compute the position of TM over time
edgePosList_TM = []
for tIdx,t in enumerate(m_MonoTM.loggedTimeVec):
    edgePosList_TM.append(m_MonoTM.xVec[np.argmax(np.abs(np.gradient(m_MonoTM.TM[:,tIdx])))])
    
# Assemble into a data frame for plotting
edgeMovementDf = pd.DataFrame({'Time':m_MonoTM.loggedTimeVec[::100], 
                               'EdgePos_TA':edgePosList_TA[::100],
                               'EdgePos_TM':edgePosList_TM[::100]})
```

```python
with sns.plotting_context("notebook", font_scale=1.7):
    plt.figure(figsize=(8,12))
    plt.scatter(x=edgeMovementDf.Time, y=edgeMovementDf.EdgePos_TA, label=r"Position of the Edge of $T_A$ in Panel A",
               marker="x", color='#FD676B',s=300)
    plt.scatter(x=edgeMovementDf.Time, y=edgeMovementDf.EdgePos_TM, label=r"Position of the Edge of $T_M$ in Panel B",
               marker="^", edgecolor="black", color='#DDDA6D',s=300)
    plt.legend(loc='lower right')
    plt.xlabel(r"$t$")
    plt.ylabel(r"$x$-Position of Tumour Front")
    # plt.title(r"Position of Tumour Front over Time")
    plt.savefig('edgePosFigure.pdf')
```

**Note:** Technically, I do see a little bit of advancement of the TM front initially, but then it regresses. This is because it is not seeded in equilibrium, and so can initially move forward before it is pushed back by the stroma. As showing and explaining this is distracts from the main story and isn't important I decided for the paper not to show this initial transient, and just plot every 100 points.

### Instability to small changes in the ICs

```python
%%time
m8 = mySolver.fullModelSolver(paramsList=paramsList, domainSizeList=domainSizeList,
                             dx=dx, dt=dt, stabiliseSpatDiscB=stabiliseSpatDiscB)
m8.Set_ICParameters(stromaParamList=[0.2,0.1],tumourAParamList=[0.1,0.1],
               tumourMParamList=[0.1,0.1],matrixParamList=[0.2,0.1],
               acidParamList=[0.1,0.1])
m8.ApplyInitialConditions()
m8.cAM=1.7
m8.cMA=1.7
```

```python
# Uncomment this for random perturbations in the ICs
# m8.TM[m8.TM[:,0]>0,0] += 1e-5*np.random.randn(np.sum(m8.TM>0))
# m8.TM[m8.TM[:,0]>1,0] = 1
# m8.TM[m8.TM[:,0]<0,0] = 0
# Use this for shifting the TM population to the right
nonZeroIdxVec = np.where(m8.TM[:,0]>0)[0]
shift = 15
m8.TM[nonZeroIdxVec[0]+shift:nonZeroIdxVec[-1]+shift+1,0] = m8.TM[nonZeroIdxVec[0]:nonZeroIdxVec[-1]+1,0]
m8.TM[:nonZeroIdxVec[0]+shift,0] = m8.TM[nonZeroIdxVec[0]+shift:nonZeroIdxVec[0]+2*shift,0][::-1]
m8.Plot([0])
```

```python
m8.Run(method=method, rtol=rtol, atol=atol, logInterval=1.)
```

```python
m8.Plot([0,m8.tEnd/2,m8.tEnd],figsize=(12,12),lineWidth=5)
plt.savefig("biStability_perturbICs.pdf")
```
